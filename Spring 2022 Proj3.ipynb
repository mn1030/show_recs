```python
recs=[]
found=False
```

```python
def printList(d):
    print("RECOMMENDATIONS-----------------------------------------------------------")
    for i in d:
        print(i)
    print("--------------------------------------------------------------------------")
```

```python
from bs4 import BeautifulSoup
import requests
import re


# DOWNLOADING TOP TV SHOW DATA
url = 'http://www.imdb.com/chart/toptv'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'lxml')

shows = soup.select('td.titleColumn')
links = [a.attrs.get('href') for a in soup.select('td.titleColumn a')]
crew = [a.attrs.get('title') for a in soup.select('td.titleColumn a')]

ratings = [b.attrs.get('data-value')
        for b in soup.select('td.posterColumn span[name=ir]')]

show_year={}
show_cast={}
show_rating={}

# eeach movie is a dictionary of its details
for index in range(0, len(shows)):
    
    
    movie_string = shows[index].get_text()
    movie = (' '.join(movie_string.split()).replace('.', ''))
    movie_title = movie[len(str(index))+1:-7]
    
    show_year[movie_title] = int(re.search('\((.*?)\)', movie_string).group(1)) #YEAR
    show_cast[movie_title] = crew[index] #CAST
    show_rating[movie_title] =float(ratings[index][:3]) #RATING
    
    recs.append({"title":movie_title,"year":show_year[movie_title],"cast":show_cast[movie_title],"rating":show_rating[movie_title]})
```

```python

#1. CHECK RATING
#remove if rating < searchRating

if found==False:
    searchRating = float(input("At least how many stars does your show have? "))
    try:
        
        for i in range(len(recs)):

            if recs[i]["rating"] < searchRating: #the shows are already sorted by rating

                recs=recs[0:i]
                break
        printList(recs)
        if (input("Did you find your show? (y/n) ") == "y"):
            found = True
            print("Enjoy the show!")
        
    except:
        print("Please type in a float")
```

```python
#2. CHECK CAST

if found == False:
    searchCast = input("Who was in your show? If you don't know, press Enter ") #might use .lower() #make lower for ease of searching
    searchCast = searchCast.lower() #get rid of case sensitivity

    if searchCast:

        #printList(recs)
        for i in range(len(recs)-1,-1,-1):
            checkCast = recs[i]["cast"].lower() #the cast names from each show already in recs
            
            if searchCast not in checkCast:
                del recs[i]
                
        printList(recs)

    if (input("Did you find your show? (y/n) ") == "y"):
        found = True
        print("Enjoy the show!")
```

```python
#3. DONE!! CHECK YEAR

if found == False:
    try:
        searchYear = int(input("What year was your show released? "))

        if searchYear:
          for i in range(len(recs)-1,-1,-1):
            if recs[i]["year"] != searchYear:
                del recs[i]

        printList(recs)

        if (input("Did you find your show? (y/n) ") == "y"):
            found = True
            print("Enjoy the show!")
    except:
        print("Please type in an integer")
```

```python
#4. CHECK TITLE


if found == False:
    searchTitle = input("What title are you looking for? If you don't know, press Enter") #might use .lower() #make lower for ease of searching
    searchTitle = searchTitle.lower() #get rid of case sensitivity

    if searchTitle:
        
        for i in range(len(recs)-1,-1,-1):
            checkTitle = recs[i]["title"].lower()

            if searchTitle not in checkTitle:
                del recs[i]
            
    printList(recs)

    if (input("Did you find your show? (y/n)") == "y"):
        found = True
        print("Enjoy the show!")


```
